### 1) Write a python program to compute distance between Given two objects represented by the tuples (22, 1, 42, 10) and (20, 0, 36, 8): 
(a) Compute the Euclidean distance between the two objects. <BR>
(b) Compute the Manhattan distance between the two objects. <BR>
(c) Compute the Minkowski distance between the two objects, using q = 3. <BR>
(d) Compute the supremum distance between the two objects.<BR>


```python
import math as m
x = (22,1,42,10)
y = (20,0,36,8)
sum = 0
for i in range(len(x)):
    sum += (x[i] - y[i])**2

print(f"Euclidean distance : {m.sqrt(sum)}")
```

```python
sum = 0 
for i in range(len(x)):
    sum += abs(x[i] - y[i])
print(f"Manhattan distance : {sum}")
```

```python
sum =  0
for i in range(len(x)):
    sum += (x[i] - y[i])**3
print(f"Minkowski distance : {(sum)**(1/3)}")
```

```python
list = []
for i in range(len(x)):
    list.append(abs(x[i]-y[i]))
print(f"Supremum distance : {max(list)}")
```

### 3) Write programs to perform the following tasks of preprocessing.

Equal Width Binning <BR>
Equal Frequency/Depth Binning

```python
data = [5,10,11,13,15,35,50,55,72,92,204,215]
df = pd.DataFrame(data, columns=['Values'])
num_bins = 3
bin_edges = np.linspace(df['Values'].min(),df['Values'].max(),num_bins+1)
df['Equal_Widht_Bin'] = pd.cut(df['Values'],bins=bin_edges,labels=[1,2,3], include_lowest=True)
print("Equal Width Binning:\n" ,df)
```

```python
no_of_data = len(data)
points_in_bin = no_of_data/num_bins

for i in range(0,len(data),int(points_in_bin)):
    print(data[i:i+int(points_in_bin)])
```

### 4)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
df = data
NewMax = 10
NewMin = 0
df['New-Min-Max-Age'] = ((df['Age'] - df['Age'].min()) / (df['Age'].max() - df["Age"].min()))*(NewMax-NewMin)+NewMin
df
```

```python
df['New-Decimal-Sacling-Age'] = df['Age']/10**len(str(int(df['Age'].max())))
df
```

```python
df['New-Z-Score-Age'] = (df['Age']-df['Age'].mean())/df['Age'].std()
df
```
